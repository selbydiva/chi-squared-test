```python
data = {

    #'age' : [15, 25, 23, 13, 19, 14, 12, 16, 20, 13, 15, 22, 20, 21, 15, 23, 14],
    'age_status' : ['teen' , 'adult', 'adult', 'teen', 'adult', 'teen', 'teen', 'teen', 'adult',
                    'teen','teen','adult','adult','adult','teen','adult','teen', 'adult', 'adult',
                    'teen', 'teen', 'teen','adult','teen', 'adult', 'teen', 'teen', 'adult','adult','adult',
                    'adult', 'adult', 'teen', 'adult', 'adult', 'teen', 'teen', 'adult', 'adult','adult','teen',
                    'teen', 'teen', 'teen', 'adult', 'adult', 'teen', 'adult', 'teen', 'adult'

                    ],
    'drink' : ['soda', 'coffee', 'soda', 'milk', 'coffee', 'soda', 'milk', 'coffee', 'coffee',
               'milk', 'soda', 'coffee', 'soda', 'coffee', 'soda', 'soda', 'milk', 'coffee', 'soda',
               'milk', 'soda', 'soda','coffee', 'soda', 'milk', 'milk', 'milk', 'coffee','soda','coffee',
               'soda', 'coffee', 'soda', 'coffee', 'soda', 'soda', 'milk', 'soda', 'coffee','soda','milk',
               'soda', 'soda', 'milk', 'coffee', 'milk', 'milk', 'coffee', 'milk','milk'
            ]

}

data = pd.DataFrame(data)
data
```

```python
import pandas as pd

data_crosstab = pd.crosstab(
		 					data['age_status'],
							data['drink'],
							margins = False)
print(data_crosstab)
```

```python
import scipy.stats as stats

chi2, p, dof, expected = stats.chi2_contingency(data_crosstab)

print('Expected Frequencies :', expected)
print('Chi Squared Value:', chi2)
print('Degrees of Freedom :', dof)
print('p-value :', p)
```
